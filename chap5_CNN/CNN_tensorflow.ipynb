```python
import tensorflow as tf
import numpy as np
#当前下载的TensorFlow是2以上版本
# TF2中没有tensorflow.examples，得改为keras，它自带MNIST数据集
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

learning_rate = 1e-4
keep_prob_rate = 0.7 # 
max_epoch = 2000
batch_size = 100

inputs = tf.keras.Input(shape=(28, 28, 1))

# 卷积层 1
x = tf.keras.layers.Conv2D(
    filters=32,
    kernel_size=(7, 7),
    strides=(1, 1),
    padding="same",
    activation="relu",
)(inputs)
x = tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=(2, 2), padding="same")(x)

# 卷积层 2
x = tf.keras.layers.Conv2D(
    filters=64,
    kernel_size=(5, 5),
    strides=(1, 1),
    padding="same",
    activation="relu",
)(x)
x = tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=(2, 2), padding="same")(x)

# 全连接层
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(1024, activation="relu")(x)
x = tf.keras.layers.Dropout(rate=1.0 - keep_prob_rate)(x)
outputs = tf.keras.layers.Dense(10, activation="softmax")(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

model.compile(
    optimizer=optimizer,
    loss="categorical_crossentropy",
    metrics=["accuracy"],
)


def compute_accuracy(v_xs, v_ys):
    preds = model.predict(v_xs, verbose=0)
    pred_cls = np.argmax(preds, axis=1)
    true_cls = np.argmax(v_ys, axis=1)
    return float(np.mean(pred_cls == true_cls))


def train_step(batch_xs, batch_ys):
    # 每一步更新一次参数
    return model.train_on_batch(batch_xs, batch_ys)


# 每100 step打印一次
for i in range(max_epoch):
    indices = np.random.randint(0, x_train.shape[0], size=batch_size)
    batch_xs = x_train[indices][..., np.newaxis]
    batch_ys = y_train[indices]

    _ = train_step(batch_xs, batch_ys)

    if i % 100 == 0:
        print(compute_accuracy(x_test[:1000, ..., np.newaxis], y_test[:1000]))
```
